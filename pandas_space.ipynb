```python
labels = ['small', 'medium', 'large', 'extra large']
ds['volume_bucket'] = pd.qcut(ds['volume'], q=len(labels), labels=labels)
```

```python
ds.head(1)
```

```python
sns.distplot(ds['carat'])
```

```python
sns.countplot(ds['cut'])
```

```python
sns.countplot(data=ds, x='cut', hue='volume_bucket')
```

```python
g = sns.JointGrid(data=ds, x='carat', y='price')
g.plot(sns.regplot, sns.displot)
```

```python
pv = ds.pivot_table(index='cut', columns= ['volume_bucket'], values='price')
```

```python
pv
```

```python
sns.heatmap(pv)
```
